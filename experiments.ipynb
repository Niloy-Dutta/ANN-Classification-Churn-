```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
#Load data set
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
# Preprocess the data 
# Drop irrelevent clumns 
data = data.drop(columns=['RowNumber','CustomerId','Surname'])
```

```python
data.head()
```

```python
data['Gender'].unique()
```

```python
# Encode Categorical variable 
label_encoder_gender = LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
data['Geography'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot_encoder_geo = OneHotEncoder(sparse_output=False)
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
# Combine one hot encoder columns with the main dataframe
data = pd.concat([data.drop(columns=['Geography']),geo_encoded_df],axis= 1)
```

```python
data.head()
```

```python
# saving the encoders in pickel file 
with open('label_encoder_gender.pkl','wb') as file: 
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb') as file: 
    pickle.dump(onehot_encoder_geo,file)
```

```python
data.head()
```

```python
# Divide the dataset into independent and dependent features 
X = data.drop(columns=['Exited'])
y = data['Exited']

#train-test split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)

```

```python
# Scale this features 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

### ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime as dt 
```

```python
(X_train.shape[1],)
```

```python
# Building the ANN Model 
model = Sequential(
    [
        Dense(64,activation='relu',input_shape=(X_train.shape[1],) ), # HL1 connected with input layer
        Dense(32,activation='relu'), # HL2 
        Dense(1,activation='sigmoid') # Output layer
    ]
)
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=.01)
loss = tf.keras.losses.BinaryCrossentropy()
```

```python
# compile the model
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
# Setup TensorBoard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir = 'logs/fit/' + dt.datetime.now().strftime('%Y%m%d-%H%M%S')
tensflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# Set up Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
# training the model 
history = model.fit(
    X_train,y_train,validation_data = (X_test,y_test),epochs = 100,
    callbacks = [tensflow_callback,early_stopping_callback]
)
```

```python
# saving the model in .h5 file
model.save('model.h5')
```
